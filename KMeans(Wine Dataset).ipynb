```python
df2=df1.drop('quality',)
df2=df1
```

```python
df2.head()
```

```python
from scipy.stats import zscore
df_scaled =df2.apply(zscore)
```

```python
df_scaled
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_range=range(1,15)
cluster_errors=[]
for num_clusters in cluster_range:
    clusters=KMeans(num_clusters)
    clusters.fit(df_scaled)
    labels=clusters.labels_
    centroids=clusters.cluster_centers_
    cluster_errors.append(clusters.inertia_)
clusters_df=pd.DataFrame({"num_clusters":cluster_range,'cluster_errors':cluster_errors})
clusters_df[0:15]
```

```python
plt.figure(figsize=(12,6))
plt.plot(clusters_df.num_clusters,clusters_df.cluster_errors,marker='o')
```

```python
kmeans = KMeans(n_clusters=6,n_init=15,random_state=2)
```

```python
kmeans.fit(df_scaled)
```

```python
centroids=kmeans.cluster_centers_
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,8))
ax=Axes3D(fig,rect=[0,0,.95,1],elev=10,azim=80)
kmeans.fit(df_scaled)
labels=kmeans.labels_
ax.scatter(df_scaled.iloc[:,0],df_scaled.iloc[:,2],df_scaled.iloc[:,4],c=labels.astype(np.float),edgecolor='k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_xlabel('Dia')
ax.set_ylabel('Mweight')
ax.set_zlabel('Ring')
ax.set_title('3D Plot of KMeans Cluster')

```
