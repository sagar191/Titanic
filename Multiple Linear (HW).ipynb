```python
x=adv_mul[['TV','Radio','Newspaper']]
y=adv_mul['Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.7,train_size=0.3,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df=pd.DataFrame(lm.coef_,x_test.columns,columns=['cofficient'])
```

```python
coeff_df
```

```python
y_pred=lm.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# Calculating Error Terms 
mse= mean_squared_error(y_test,y_pred)
r_squared=r2_score(y_test,y_pred)
```

```python
print(mse)
print(r_squared) # r_squared value 0.8 more than model is good.
```
