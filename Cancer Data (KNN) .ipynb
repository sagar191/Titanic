```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
l=load_breast_cancer() #storing the object in L variable
```

```python
cancer=pd.DataFrame(l.data,columns=l.feature_names)
cancer
```

```python
l.target_names      #Malignant=0,Benign=1
```

```python
# import ML Library -KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier() #initiating the Machine
```

```python
# Train and Test
#(x)predictors:cancer dataset
#(y)target -cancer.target
from sklearn.model_selection import train_test_split
```

```python
# Defining the x and y
x=cancer.drop('target',axis=1)
y=cancer.target
```

```python
# splitting the data in train and test 
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=123)
```

```python
xtrain.head() #
```

```python
ytrain.head()
```

```python
# create the model
knn.fit(xtrain,ytrain)
```

```python
#make the prediction
pred=knn.predict(xtest)
```

```python
# Import the Metrics
from sklearn.metrics import accuracy_score,cohen_kappa_score
```

```python
#calc the accuracy of the model
accuracy_score(ytest,pred)
```
